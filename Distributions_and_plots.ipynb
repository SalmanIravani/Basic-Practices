```python
%%markdown

## Binomial Distribution

The **Binomial distribution** is a discrete probability distribution that describes the number of successes in a fixed number of independent Bernoulli trials. A Bernoulli trial is a single experiment with only two possible outcomes: success or failure.

The two main parameters of the Binomial distribution are:

*   **n:** The number of independent trials.
*   **p:** The probability of success on a single trial.

The number of successes in *n* trials follows a Binomial distribution, often denoted as $B(n, p)$. The probability mass function (PMF) of a Binomial distribution is given by:

$P(X=k) = \binom{n}{k} p^k (1-p)^{n-k}$

where $X$ is the number of successes, $k$ is the specific number of successes, and $\binom{n}{k}$ is the binomial coefficient, calculated as $\frac{n!}{k!(n-k)!}$.

**Use Cases:**

The Binomial distribution is widely used to model situations involving a series of independent trials with binary outcomes, such as:

*   The number of heads in a fixed number of coin flips.
*   The number of defective items in a sample from a production line.
*   The number of successful treatments in a clinical trial with a fixed number of patients.
*   The number of voters who support a particular candidate in a random sample.
```

```python
%%markdown

## Poisson Distribution

The **Poisson distribution** is a discrete probability distribution that expresses the probability of a given number of events occurring in a fixed interval of time or space if these events occur with a known constant mean rate and independently of the time since the last event.

The Poisson distribution has a single parameter:

*   ** (lambda):** The average number of events in the given interval. Lambda must be a positive real number.

The probability mass function (PMF) of a Poisson distribution is given by:

$P(X=k) = \frac{\lambda^k e^{-\lambda}}{k!}$

where $X$ is the number of events, $k$ is the actual number of events (a non-negative integer), $e$ is the base of the natural logarithm ($e \approx 2.71828$), and $k!$ is the factorial of $k$.

**Use Cases:**

The Poisson distribution is commonly used to model the number of times an event occurs in a fixed interval, such as:

*   The number of customers arriving at a store per hour.
*   The number of defects per unit area of a manufactured material.
*   The number of traffic accidents at a specific intersection per month.
*   The number of emails received in a day.
*   The number of radioactive decays per second.
```

```python
%%markdown

## Normal Distribution

The **Normal distribution**, also known as the Gaussian distribution or bell curve, is a continuous probability distribution that is symmetric about its mean. It is arguably the most important distribution in statistics because many natural phenomena follow this distribution, and it plays a central role in the Central Limit Theorem.

The Normal distribution is defined by two parameters:

*   ** (mu):** The mean (average) of the distribution. This determines the center of the distribution.
*   ** (sigma):** The standard deviation of the distribution. This determines the spread or width of the distribution. A smaller standard deviation means the data is clustered closer to the mean, while a larger standard deviation indicates the data is more spread out.

The probability density function (PDF) of a Normal distribution is given by:

$f(x; \mu, \sigma) = \frac{1}{\sigma \sqrt{2\pi}} e^{-\frac{1}{2}\left(\frac{x-\mu}{\sigma}\right)^2}$

The shape of the Normal distribution is characterized by its distinctive bell curve: it is symmetric, unimodal (has a single peak), and its tails extend infinitely in both directions but approach the x-axis without ever touching it.

**Use Cases:**

The Normal distribution is widely used to model various phenomena in nature and science, including:

*   Heights, weights, and other physical characteristics of a population.
*   Measurement errors in experiments.
*   Scores on standardized tests (e.g., IQ scores).
*   Financial data, such as stock returns (though often with some caveats).
*   The distribution of sample means (due to the Central Limit Theorem).
```

## Explain skewness

### Subtask:
Describe what skewness is, the different types (positive, negative, zero), and how it relates to the shape of a distribution.

```python
%%markdown

## Skewness

**Skewness** is a measure of the asymmetry of the probability distribution of a real-valued random variable about its mean. In simpler terms, it indicates the degree to which the distribution of a dataset is "tilted" or "lopsided" to one side.

There are three main types of skewness:

1.  **Positive Skew (Right-Skewed):**
    *   In a positively skewed distribution, the tail of the distribution is longer on the right side.
    *   This indicates that there are more extreme values (outliers) on the higher end of the data.
    *   The peak of the distribution is shifted to the left.
    *   Typically, for a positively skewed distribution, the mean is greater than the median, and the median is often greater than the mode (Mean > Median > Mode).

2.  **Negative Skew (Left-Skewed):**
    *   In a negatively skewed distribution, the tail of the distribution is longer on the left side.
    *   This indicates that there are more extreme values (outliers) on the lower end of the data.
    *   The peak of the distribution is shifted to the right.
    *   Typically, for a negatively skewed distribution, the mean is less than the median, and the median is often less than the mode (Mean < Median < Mode).

3.  **Zero Skew:**
    *   A distribution has zero skew if it is perfectly symmetric around its mean.
    *   The left and right sides of the distribution are mirror images of each other.
    *   The most well-known example of a distribution with zero skew is the Normal (Gaussian) distribution.
    *   In a perfectly symmetric distribution, the mean, median, and mode are all equal (Mean = Median = Mode).

**Relationship to Distribution Shape:**

Skewness directly affects the visual shape of a histogram or probability distribution curve.

*   A **positively skewed** distribution will appear to have a peak shifted towards the left, with a long "tail" extending towards the right along the x-axis.
*   A **negatively skewed** distribution will have its peak shifted towards the right, with a long "tail" extending towards the left along the x-axis.
*   A **zero-skewed** distribution will be balanced, with its peak in the center and tails of roughly equal length on both sides, appearing symmetric.

Understanding skewness is important in data analysis as it can indicate potential issues with the data's representativeness, inform the choice of statistical methods (many methods assume normality, which implies zero skew), and provide insights into the underlying process generating the data.
```

## Calculate and visualize skewness

### Subtask:
Provide Python code using `scipy.stats` to calculate skewness and use `matplotlib.pyplot` to visualize examples of skewed distributions.

Provide Python code using `scipy.stats` to calculate skewness and use `matplotlib.pyplot` to visualize examples of skewed distributions as requested in the subtask.



```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import skew

# Generate three datasets with different skewness characteristics
# Positively skewed (using Exponential distribution)
pos_skew_data = np.random.exponential(scale=1.0, size=100000)

# Negatively skewed (using Beta distribution with alpha > beta, e.g., alpha=5, beta=1)
neg_skew_data = np.random.beta(a=5, b=1, size=100000)

# Roughly symmetric (using Normal distribution)
sym_data = np.random.normal(loc=0.0, scale=1.0, size=100000)

# Calculate skewness for each dataset
skew_pos = skew(pos_skew_data)
skew_neg = skew(neg_skew_data)
skew_sym = skew(sym_data)

# Print the calculated skewness values
print(f"Skewness of Positively Skewed Data: {skew_pos:.2f}")
print(f"Skewness of Negatively Skewed Data: {skew_neg:.2f}")
print(f"Skewness of Symmetric Data: {skew_sym:.2f}")

# Create a figure and subplots
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Plot histogram for positively skewed data
axes[0].hist(pos_skew_data, bins=100, density=True, alpha=0.7, color='skyblue')
axes[0].set_title(f'Positively Skewed (Skew = {skew_pos:.2f})')
axes[0].set_xlabel('Value')
axes[0].set_ylabel('Probability Density')

# Plot histogram for negatively skewed data
axes[1].hist(neg_skew_data, bins=100, density=True, alpha=0.7, color='orange')
axes[1].set_title(f'Negatively Skewed (Skew = {skew_neg:.2f})')
axes[1].set_xlabel('Value (Probability/Proportion)')
axes[1].set_ylabel('Probability Density')

# Plot histogram for symmetric data
axes[2].hist(sym_data, bins=100, density=True, alpha=0.7, color='salmon')
axes[2].set_title(f'Roughly Symmetric (Skew = {skew_sym:.2f})')
axes[2].set_xlabel('Value')
axes[2].set_ylabel('Probability Density')

# Add a main title to the figure
fig.suptitle('Examples of Skewed Distributions', fontsize=16)

# Adjust layout
plt.tight_layout(rect=[0, 0.03, 1, 0.95]) # Adjust layout to make space for the main title

# Display the plot
plt.show()
```

```python
%%markdown

## Kurtosis

**Kurtosis** is a measure that describes the shape of the tails and the peakedness of a probability distribution relative to the tails and peakedness of a Normal distribution. It tells us how much of the data is concentrated in the tails vs. in the center (peak) of the distribution.

Kurtosis is often reported as **excess kurtosis**, which is calculated as Kurtosis - 3. The Normal distribution has a kurtosis of 3, so its excess kurtosis is 0. Using excess kurtosis makes the Normal distribution a baseline for comparison.

There are three main types of kurtosis:

1.  **Mesokurtic:**
    *   A distribution that has a kurtosis equal to that of a Normal distribution (Kurtosis = 3, Excess Kurtosis = 0).
    *   It has a moderate peak and moderate tails. The Normal distribution is the defining example of a mesokurtic distribution.

2.  **Leptokurtic:**
    *   A distribution that has fatter tails and a sharper, higher peak than a Normal distribution (Kurtosis > 3, Excess Kurtosis > 0).
    *   "Lepto" means thin, but in the context of tails, it refers to the distribution having more probability mass in the tails than a Normal distribution. This also often results in a taller peak.
    *   This indicates a higher probability of extreme values (outliers) compared to a Normal distribution.

3.  **Platykurtic:**
    *   A distribution that has thinner tails and a flatter, lower peak than a Normal distribution (Kurtosis < 3, Excess Kurtosis < 0).
    *   "Platy" means broad or flat.
    *   This indicates a lower probability of extreme values compared to a Normal distribution; the data is more spread out evenly throughout the distribution rather than concentrated in the tails or a sharp peak.

**Relationship to Tails and Peak:**

Kurtosis is primarily a measure of the "tailedness" of a distribution, but it also influences the peak:

*   **Higher Kurtosis (Leptokurtic):** Implies fatter tails (more outliers) and often a higher, sharper peak to compensate for the probability mass in the tails (as the total area under the curve must equal 1).
*   **Lower Kurtosis (Platykurtic):** Implies thinner tails (fewer outliers) and a flatter, broader peak because the probability mass is more spread out towards the shoulders of the distribution.

In summary, kurtosis helps us understand the concentration of data in the tails and center of a distribution, providing valuable information about the likelihood of extreme values.
```

## Calculate and visualize kurtosis

### Subtask:
Provide Python code using `scipy.stats` to calculate kurtosis and use `matplotlib.pyplot` to visualize examples of distributions with different kurtosis.


**Reasoning**:
Generate data from distributions with different kurtosis, calculate their kurtosis, and plot their histograms in subplots with titles showing the calculated kurtosis values.



```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import kurtosis, laplace, uniform, norm

# Generate three datasets with different kurtosis characteristics
# Leptokurtic distribution (e.g., Laplace distribution)
# The kurtosis of a Laplace distribution is 6, excess kurtosis is 3.
leptokurtic_data = laplace.rvs(size=100000)

# Mesokurtic distribution (e.g., Normal distribution)
# The kurtosis of a Normal distribution is 3, excess kurtosis is 0.
mesokurtic_data = norm.rvs(size=100000)

# Platykurtic distribution (e.g., Uniform distribution)
# The kurtosis of a Uniform distribution is 1.8, excess kurtosis is -1.2.
platykurtic_data = uniform.rvs(loc=0, scale=10, size=100000)

# Calculate excess kurtosis for each dataset
# Set fisher=True for excess kurtosis (kurtosis - 3)
kurtosis_lepto = kurtosis(leptokurtic_data, fisher=True)
kurtosis_meso = kurtosis(mesokurtic_data, fisher=True)
kurtosis_platy = kurtosis(platykurtic_data, fisher=True)

# Print the calculated excess kurtosis values
print(f"Excess Kurtosis of Leptokurtic Data (Laplace): {kurtosis_lepto:.2f}")
print(f"Excess Kurtosis of Mesokurtic Data (Normal): {kurtosis_meso:.2f}")
print(f"Excess Kurtosis of Platykurtic Data (Uniform): {kurtosis_platy:.2f}")

# Create a figure and subplots
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Plot histogram for leptokurtic data
axes[0].hist(leptokurtic_data, bins=100, density=True, alpha=0.7, color='teal')
axes[0].set_title(f'Leptokurtic Distribution\n(Excess Kurtosis = {kurtosis_lepto:.2f})')
axes[0].set_xlabel('Value')
axes[0].set_ylabel('Probability Density')

# Plot histogram for mesokurtic data
axes[1].hist(mesokurtic_data, bins=100, density=True, alpha=0.7, color='orange')
axes[1].set_title(f'Mesokurtic Distribution\n(Excess Kurtosis = {kurtosis_meso:.2f})')
axes[1].set_xlabel('Value')
axes[1].set_ylabel('Probability Density')

# Plot histogram for platykurtic data
axes[2].hist(platykurtic_data, bins=100, density=True, alpha=0.7, color='purple')
axes[2].set_title(f'Platykurtic Distribution\n(Excess Kurtosis = {kurtosis_platy:.2f})')
axes[2].set_xlabel('Value')
axes[2].set_ylabel('Probability Density')

# Add a main title to the figure
fig.suptitle('Examples of Distributions with Different Kurtosis', fontsize=16)

# Adjust layout
plt.tight_layout(rect=[0, 0.03, 1, 0.95]) # Adjust layout to make space for the main title

# Display the plot
plt.show()
```
